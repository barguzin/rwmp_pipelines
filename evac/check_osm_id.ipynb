```python
import geopandas as gpd
import osmnx as ox 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# get OSM data for bbox area 
# read shapefile for studya area 
sa = gpd.read_file('C:/Users/barguzin/Documents/Github/rwmp_pipelines/roadsevac_demo2_epsg26910.geojson')
print(sa.crs)
print(sa.shape)
sa.plot()
```

```python
sa.to_crs('epsg:4326', inplace=True)
print(sa.crs)
```

```python
poly = sa.geometry.iloc[0]

G = ox.graph_from_polygon(poly, network_type="drive_service") # include drive and service roads

fig, ax = ox.plot_graph(G, node_size=1, edge_color='w', edge_linewidth=.2)
```
